```python
import os 
import h5py
import numpy as np
```

```python
# -- plotting -- 
import matplotlib as mpl
import matplotlib.pyplot as plt
#mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['axes.linewidth'] = 1.5
mpl.rcParams['axes.xmargin'] = 1
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 1.5
mpl.rcParams['ytick.labelsize'] = 'x-large'
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.major.width'] = 1.5
mpl.rcParams['legend.frameon'] = False
```

```python
fbgs = os.path.join('/global/cfs/cdirs/desi/mocks/bgs/MXXL/desi_footprint/v0.0.4', 'BGS.hdf5')
```

```python
bgs = h5py.File(fbgs, 'r')
```

```python
for k in bgs['Header'].keys(): 
    print(k, bgs['Header'][k][...])
```

```python
mock = bgs['Data']
```

```python
fig = plt.figure(figsize=(10,5))
sub = fig.add_subplot(111)
sub.scatter((mock['ra'][...][::100]-80) % 360 + 80, mock['dec'][...][::100], c='k', s=1)
sub.set_xlabel('RA', fontsize=20)
sub.set_xlim(80, 440)
sub.set_ylabel('Dec', fontsize=20)
```

```python
fig = plt.figure(figsize=(6,5))
sub = fig.add_subplot(111)
sub.hist(mock['z_obs'][...], range=(0., 1.), bins=80, density=True)
sub.set_xlabel('redshift', fontsize=20)
sub.set_xlim(0., 1.)
```
